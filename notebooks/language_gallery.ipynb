# universal-render — language gallery

Every language rendered one by one, grouped by script family. Each language shows its **autonym** (its own name in its own script) and a common greeting. Running the notebook also saves each render to `gallery_output/<Family>/<Language>.png`, so you get one folder per family with one image per language.

If you read one of these languages, please check that its line is correct and let us know at mrinalbasak23@gmail.com.

## Setup

```python
%pip -q install --upgrade universal-render
import platform, subprocess
if platform.system() == 'Linux':
    subprocess.run('apt-get -qq update', shell=True, capture_output=True)
    subprocess.run(
        'apt-get -qq install -y fonts-noto-core fonts-noto-ui-core '
        'fonts-noto-cjk libegl1 libgl1 libxkbcommon0 libdbus-1-3',
        shell=True, capture_output=True)
print('setup done')
```

```python
import os, platform
# offscreen is required on headless Linux, and it BREAKS text rendering
# on Windows and macOS, where the native plugin works without a window.
if platform.system() == 'Linux':
    os.environ.setdefault('QT_QPA_PLATFORM', 'offscreen')
import matplotlib.pyplot as plt
import universal_render as ur
ur.init_renderer()

def show_language(language, script, text, family):
    outdir = os.path.join('gallery_output', family)
    os.makedirs(outdir, exist_ok=True)
    fig, ax = plt.subplots(figsize=(9, 1.5))
    ax.axis('off')
    ur.multilingual_text(ax, 0.01, 0.5, text, coord='axes',
                         ha='left', va='center', font_size=40)
    ax.text(0.99, 0.5, f'{language}  ({script})', transform=ax.transAxes,
            ha='right', va='center', fontsize=11, color='0.45')
    fig.savefig(os.path.join(outdir, f'{language}.png'),
                dpi=150, bbox_inches='tight')
    plt.show()
```
