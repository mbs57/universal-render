# universal-render — per-language figure set

For every language this notebook generates **two fully localized figures**, a confusion matrix and a bar chart, and saves them to `figures_output/<Language>/`. You get one folder per language, each holding `confusion_matrix.png` and `bar_chart.png`.

To keep every figure verifiably correct without relying on machine translation, each title is the language's **autonym** (its own name in its own script) and all numeric labels use that script's **native digits**. So the title exercises real shaping and the data exercises native-numeral rendering, with nothing that could be a mistranslation.

```python
import os, platform
# offscreen is required on headless Linux, and it BREAKS text rendering
# on Windows and macOS, where the native plugin works without a window.
if platform.system() == 'Linux':
    os.environ.setdefault('QT_QPA_PLATFORM', 'offscreen')
import numpy as np
import matplotlib.pyplot as plt
import universal_render as ur
ur.init_renderer()

def native(n, script):
    return ur.to_native_numerals(n, script)

def make_confusion(language, script, autonym):
    outdir = os.path.join('figures_output', language)
    os.makedirs(outdir, exist_ok=True)
    cm = np.array([[45, 3, 2], [4, 42, 4], [1, 5, 44]], float)
    classes = [native(i + 1, script) for i in range(3)]
    cells = np.array([[native(int(v), script) for v in row] for row in cm],
                     dtype=object)
    fig, ax = plt.subplots(figsize=(5, 4.5))
    ur.multilingual_heatmap(ax, cm, labels=cells,
                            row_labels=classes, col_labels=classes,
                            title=autonym, cell_font_size=36,
                            tick_font_size=28, title_font_size=34,
                            colorbar=True,
                            imshow_kwargs={'cmap': 'Blues'})
    ur.apply_multilingual_layout(fig, auto=True)
    fig.savefig(os.path.join(outdir, 'confusion_matrix.png'),
                dpi=150, bbox_inches='tight')
    plt.show()

def make_barchart(language, script, autonym):
    outdir = os.path.join('figures_output', language)
    os.makedirs(outdir, exist_ok=True)
    cats, vals = [1, 2, 3, 4], [4, 7, 5, 6]
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.bar(cats, vals, color='#4477aa')
    ur.set_multilingual_title(ax, autonym, font_size=34)
    ur.set_multilingual_xticks(ax, cats, [native(c, script) for c in cats],
                               font_size=28)
    ur.set_multilingual_numeric_ticks(ax, script=script, axis='y',
                                      positions=[0, 2, 4, 6, 8], font_size=26)
    ur.apply_multilingual_layout(fig, auto=True)
    fig.savefig(os.path.join(outdir, 'bar_chart.png'),
                dpi=150, bbox_inches='tight')
    plt.show()

def render_language(language, script, autonym):
    print(f'--- {language} ({script}) ---')
    make_confusion(language, script, autonym)
    make_barchart(language, script, autonym)
```
