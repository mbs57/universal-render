# universal-render — per-language figure set

For every language this notebook generates **two fully localized figures** and saves them to `figures_output/<Language>/`, giving one folder per language:

1. **`confusion_matrix.png`** demonstrates **numerals**. Where a script has a living positional digit system (all the Indic scripts, Arabic, Persian, Thai, Lao, Khmer, Myanmar, Tibetan) the counts appear in native digits. For Greek, Cyrillic, Hebrew, CJK and the rest, 0-9 is the correct modern convention and is used.
2. **`bar_chart.png`** demonstrates **words**. The bars are labelled with the number words one, two, three and four in that language, so conjuncts, joining and combining marks are all exercised.

Titles are each language's **autonym**, its own name in its own script.

⚠️ **If you read one of these languages, please check its number words.** They were compiled by hand and any correction is welcome at mrinalbasak23@gmail.com or via a GitHub issue.

```python
import os, platform
# offscreen is required on headless Linux, and it BREAKS text rendering
# on Windows and macOS, where the native plugin works without a window.
if platform.system() == 'Linux':
    os.environ.setdefault('QT_QPA_PLATFORM', 'offscreen')
import numpy as np
import matplotlib.pyplot as plt
import universal_render as ur
ur.init_renderer()

def native(n, script):
    return ur.to_native_numerals(n, script)

def make_confusion(language, script, autonym):
    """Confusion matrix: demonstrates native numerals."""
    outdir = os.path.join('figures_output', language)
    os.makedirs(outdir, exist_ok=True)
    cm = np.array([[45, 3, 2], [4, 42, 4], [1, 5, 44]], float)
    classes = [native(i + 1, script) for i in range(3)]
    cells = np.array([[native(int(v), script) for v in row] for row in cm],
                     dtype=object)
    fig, ax = plt.subplots(figsize=(5, 4.5))
    ur.multilingual_heatmap(ax, cm, labels=cells,
                            row_labels=classes, col_labels=classes,
                            title=autonym, cell_font_size=36,
                            tick_font_size=28, title_font_size=34,
                            colorbar=True,
                            imshow_kwargs={'cmap': 'Blues'})
    ur.apply_multilingual_layout(fig, auto=True)
    fig.savefig(os.path.join(outdir, 'confusion_matrix.png'),
                dpi=150, bbox_inches='tight')
    plt.show()

def make_barchart(language, script, autonym, words):
    """Horizontal bar chart: demonstrates real words in the script."""
    outdir = os.path.join('figures_output', language)
    os.makedirs(outdir, exist_ok=True)
    pos, vals = [1, 2, 3, 4], [4, 7, 5, 6]
    fig, ax = plt.subplots(figsize=(7, 4.4))
    ax.barh(pos, vals, color='#4477aa')
    ax.set_yticks(pos)
    ax.invert_yaxis()
    ur.set_multilingual_title(ax, autonym, font_size=34)
    ur.set_multilingual_yticks(ax, pos, words, font_size=30)
    ur.set_multilingual_numeric_ticks(ax, script=script, axis='x',
                                      positions=[0, 2, 4, 6, 8],
                                      font_size=26)
    ur.apply_multilingual_layout(fig, auto=True)
    fig.savefig(os.path.join(outdir, 'bar_chart.png'),
                dpi=150, bbox_inches='tight')
    plt.show()

def render_language(language, script, autonym, words):
    print(f'--- {language} ({script}) ---')
    make_confusion(language, script, autonym)
    make_barchart(language, script, autonym, words)
```
